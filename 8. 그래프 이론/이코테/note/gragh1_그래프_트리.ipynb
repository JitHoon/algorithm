#### `그래프`
- 노드와 노드 사이에 연결된 간선 정보를 가지고 있는 자료구조.
- 서로 다른 객체가 연결되어 있는 문제에서 활용
- 구현 방법 1. `인접 리스트`(리스트 사용, 필요 메모리 O(`E`))
- 구현 방법 2. `인접 행렬`(2차원 배열, 필요 메모리 : O(`V^2`)) (V: 노드 개수, E: 간선 개수)

- 방향성 : 방향 or 무방향 그래프
- 순환성 : 순환 및 비순환
- 루트 노드 존재 여부 : 없음
- 노드간 관계성 : 없음
- 모델의 종류 : 네트워크 모델

#### `그래프_트리`
- 방향성 : 방향 그래프
- 순환성 : 비순환
- 루트 노드 존재 여부 : 로트 노드 존재
- 노드간 관계성 : 부모와 자식 관계
- 모델의 종류 : 계층 모델
